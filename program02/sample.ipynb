2. Write funtion that finds the 50 most frequent words in a text that are not in the stop words.

```python
from nltk.corpus import stopwords
from nltk import FreqDist
from nltk.tokenize import word_tokenize
```

```python
def most_frequent_words(text):
    words=word_tokenize(text)
    stop_word=set(stopwords.words('english'))
    filtered_words=[word.lower() for word in words if word.isalpha() and word.lower() not in stop_word]
    freq_dist=FreqDist(filtered_words)
    most_frequent=freq_dist.most_common(50)
    return most_frequent
```

```python
text=open('sample.txt').read()
result=most_frequent_words(text)
print(result)
```
